### Iris Dataset
This is perhaps the best known database to be found in the pattern recognition literature. Fisher's paper is a classic in the field and is referenced frequently to this day. (See Duda & Hart, for example.) The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.

### Apply vs Iterating

```python
df.iloc[:, :4].apply(np.sqrt)
```

```python
df.drop(["class"], axis=1) + 2
```

```python
%%timeit
new_df = pd.DataFrame()
for i in range(df.shape[0]):
    new_df = pd.concat([new_df, pd.DataFrame(df.drop(["class"], axis=1).iloc[i] + 2).T], axis=0)
new_df
```

```python
def plustwo(a):
    return a + 2

df.iloc[:, :4].apply(lambda a: a + 2)
df.iloc[:, :4].apply(plustwo)
```

```python
%%timeit
df.iloc[:, :4].apply(plustwo)
```

### Scatter

```python
x = 'sepal length'
y = 'petal length'
colors = {'Iris-setosa':'blue', 'Iris-versicolor':'red', 'Iris-virginica':'green'}  # C0, C1 ... are the default plt colors
df.plot.scatter(x, y, c=df["class"].map(colors), )

plt.show()
```

```python
x = 'sepal length'
y = 'petal length'

vals = list(df['class'].unique())

for val in vals:
    df_val = df[df['class']==val]
    plt.scatter(df_val[x], df_val[y], s=50, edgecolor='black', label=val, marker = '^')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.xlabel(x, fontsize=25)
plt.ylabel(y, fontsize=20)

plt.grid(alpha=0.5)
plt.legend(loc='best')
plt.show()
```

```python
sns.scatterplot(data=df, x="sepal length", y="petal length", hue="class", style="class")
```

```python
pd.plotting.scatter_matrix(df, c=df["class"].map(colors), figsize=(8,8))
plt.show()
```

```python
sns.pairplot(data=df, hue="class", markers=["o", "s", "D"])
plt.show()
```
